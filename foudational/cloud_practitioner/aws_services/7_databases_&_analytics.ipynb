Databases & Analytics

Introdução a Bancos de Dados

    • Armazenar dados em disco (EFS, EBS, Armazenamento de Instância EC2, S3) pode ter seus limites
    • Às vezes, você deseja armazenar dados em um banco de dados...
    • Você pode estruturar os dados
    • Você cria índices para consultar / pesquisar eficientemente os dados
    • Você define relacionamentos entre seus conjuntos de dados
    • Bancos de dados são otimizados para um propósito e vêm com diferentes características, formatos e restrições


Bancos de Dados Relacionais

    • Parece exatamente como planilhas do Excel, com links entre elas!
    • Pode-se usar a linguagem SQL para realizar consultas / buscas




Bancos de Dados NoSQL

    • NoSQL = não-SQL = bancos de dados não relacionais
    • Bancos de dados NoSQL são construídos para modelos de dados específicos e têm esquemas flexíveis para construir aplicações modernas.
    • Benefícios:
    • Flexibilidade: fácil de evoluir o modelo de dados
    • Escalabilidade: projetado para escalar por meio de clusters distribuídos
    • Alta performance: otimizado para um modelo de dados específico
    • Altamente funcional: tipos otimizados para o modelo de dados
    • Exemplos: bancos de dados de chave-valor, de documentos, de grafo, em memória, de busca

Exemplo de Dados NoSQL: JSON

    • JSON = JavaScript Object Notation
    • JSON é uma forma comum de dados que se encaixa em um modelo NoSQL
    • Os dados podem ser aninhados
    • Os campos podem mudar ao longo do tempo
    • Suporte para novos tipos: arrays, etc...




Bancos de Dados e Responsabilidade Compartilhada na AWS

    • A AWS oferece uso para gerenciar diferentes bancos de dados
    • Os benefícios incluem:
    • Provisionamento Rápido, Alta Disponibilidade, Escalonamento Vertical e Horizontal
    • Backup e Restauração Automatizados, Operações, Atualizações
    • O Patching do Sistema Operacional é tratado pela AWS
    • Monitoramento, alerta
    • Observação: muitas tecnologias de bancos de dados podem ser executadas na EC2, mas você deve lidar sozinho com a resiliência, backup, patching, alta disponibilidade, tolerância a falhas, escalonamento...

Visão Geral do AWS RDS

    • RDS significa Relational Database Service
    • É um serviço de banco de dados gerenciado para uso de banco de dados que utilizam SQL como linguagem de consulta.
    • Ele permite que você crie bancos de dados na nuvem que são gerenciados pela AWS
    • Postgres
    • MySQL
    • MariaDB
    • Oracle
    • Microsoft SQL Server
    • Aurora (Banco de dados proprietário da AWS)

Vantagens de usar RDS em comparação com implantar um banco de dados em EC2:

    • RDS é um serviço gerenciado:
    • Provisionamento automatizado, patching do sistema operacional
    • Backups contínuos e restauração para um timestamp específico (Ponto na Restauração de Tempo)!
    • Painéis de monitoramento
    • Réplicas de leitura para melhorar o desempenho de leitura
    • Configuração de Múltiplas Zonas de Disponibilidade para DR (Recuperação de Desastres)
    • Janelas de manutenção para atualizações
    • Capacidade de escalonamento (vertical e horizontal)
    • Armazenamento suportado pelo EBS (gp2 ou io1)
    • MAS você não pode SSH em suas instâncias




Amazon Aurora

    • Aurora é uma tecnologia proprietária da AWS (não open source)
    • PostgreSQL e MySQL são ambos suportados como banco de dados do Aurora
    • Aurora é "otimizado para nuvem da AWS" e alega melhoria de desempenho de 5x sobre MySQL no RDS, mais de 3x o desempenho do Postgres no RDS
    • O armazenamento do Aurora cresce automaticamente em incrementos de 10GB, até 128 TB
    • Aurora custa mais do que RDS (20% a mais) - mas é mais eficiente
    • Não está incluído no nível gratuito



Amazon Aurora Serverless

    • Instantiação automatizada do banco de dados e dimensionamento automático com base no uso real
    • PostgreSQL e MySQL são ambos suportados como banco de dados do Aurora Serverless
    • Não há necessidade de planejamento de capacidade
    • Menor sobrecarga de gerenciamento
    • Pagamento por segundo, pode ser mais econômico
    • Casos de uso: bom para cargas de trabalho pouco frequentes, intermitentes ou imprevisíveis...




Você pode criar snapshots do seu RDS, assim você pode aumentar a potencia dele, migrar para outra região, etc....

Visão Geral do Amazon ElastiCache

    Da mesma forma que o RDS é para obter bancos de dados relacionais gerenciados...
    ElastiCache é para obter Redis ou Memcached gerenciados.
    Caches são bancos de dados em memória com alto desempenho e baixa latência.
    Ajuda a reduzir a carga nos bancos de dados para cargas de trabalho intensivas de leitura.
    A AWS cuida da manutenção / patching do sistema operacional, otimizações, configuração, monitoramento, recuperação de falhas e backups.

    

DynamoDB

    • Totalmente Gerenciado Altamente disponível com replicação em 3 Zonas de Disponibilidade
    • Banco de dados NoSQL - não é um banco de dados relacional
    • Escala para cargas de trabalho massivas, banco de dados distribuído "serverless"
    • Milhões de solicitações por segundo, trilhões de linhas, centenas de TB de armazenamento
    • Rápido e consistente em desempenho
    • Latência de milissegundos de um dígito - recuperação de baixa latência
    • Integrado com IAM para segurança, autorização e administração
    • Baixo custo e capacidades de escalonamento automático
    • Classe de Tabela de Acesso Padrão e de Acesso Infrequente (IA)

Visão Geral do Redshift

    O Redshift é baseado no PostgreSQL, mas não é usado para OLTP.
    É OLAP - processamento analítico online (análises e data warehousing).
    Carrega dados uma vez a cada hora, não a cada segundo.
    Desempenho 10 vezes melhor do que outros data warehouses, escalável para petabytes de dados.
    Armazenamento de dados por colunas (em vez de baseado em linhas).
    Execução de Consulta em Paralelo Massivamente (MPP), altamente disponível.
    Pague conforme o uso com base nas instâncias provisionadas.
    Possui uma interface SQL para realizar as consultas.
    Ferramentas de BI como AWS Quicksight ou Tableau se integram com ele.

Redshift Serverless

    Provisiona e dimensiona automaticamente a capacidade subjacente do data warehouse.
    Execute cargas de trabalho analíticas sem gerenciar a infraestrutura do data warehouse.
    Pague apenas pelo que usar (economize custos).
    Casos de uso: Relatórios, painéis de controle, análises em tempo real...



Amazon EMR

    EMR significa "Elastic MapReduce".
    EMR ajuda a criar clusters Hadoop (Big Data) para analisar e processar vastas quantidades de dados.
    Os clusters podem ser compostos por centenas de instâncias EC2.
    Também suporta Apache Spark, HBase, Presto, Flink...
    EMR cuida de toda a provisão e configuração.
    Escalonamento automático e integrado com instâncias Spot.
    Casos de uso: processamento de dados, aprendizado de máquina, indexação web, big data.
