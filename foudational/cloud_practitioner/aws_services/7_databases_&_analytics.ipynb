Databases & Analytics

Introduo a Bancos de Dados

     Armazenar dados em disco (EFS, EBS, Armazenamento de Instncia EC2, S3) pode ter seus limites
     s vezes, voc deseja armazenar dados em um banco de dados...
     Voc pode estruturar os dados
     Voc cria ndices para consultar / pesquisar eficientemente os dados
     Voc define relacionamentos entre seus conjuntos de dados
     Bancos de dados so otimizados para um propsito e vm com diferentes caractersticas, formatos e restries


Bancos de Dados Relacionais

     Parece exatamente como planilhas do Excel, com links entre elas!
     Pode-se usar a linguagem SQL para realizar consultas / buscas




Bancos de Dados NoSQL

     NoSQL = no-SQL = bancos de dados no relacionais
     Bancos de dados NoSQL so construdos para modelos de dados especficos e tm esquemas flexveis para construir aplicaes modernas.
     Benefcios:
     Flexibilidade: fcil de evoluir o modelo de dados
     Escalabilidade: projetado para escalar por meio de clusters distribudos
     Alta performance: otimizado para um modelo de dados especfico
     Altamente funcional: tipos otimizados para o modelo de dados
     Exemplos: bancos de dados de chave-valor, de documentos, de grafo, em memria, de busca

Exemplo de Dados NoSQL: JSON

     JSON = JavaScript Object Notation
     JSON  uma forma comum de dados que se encaixa em um modelo NoSQL
     Os dados podem ser aninhados
     Os campos podem mudar ao longo do tempo
     Suporte para novos tipos: arrays, etc...




Bancos de Dados e Responsabilidade Compartilhada na AWS

     A AWS oferece uso para gerenciar diferentes bancos de dados
     Os benefcios incluem:
     Provisionamento Rpido, Alta Disponibilidade, Escalonamento Vertical e Horizontal
     Backup e Restaurao Automatizados, Operaes, Atualizaes
     O Patching do Sistema Operacional  tratado pela AWS
     Monitoramento, alerta
     Observao: muitas tecnologias de bancos de dados podem ser executadas na EC2, mas voc deve lidar sozinho com a resilincia, backup, patching, alta disponibilidade, tolerncia a falhas, escalonamento...

Viso Geral do AWS RDS

     RDS significa Relational Database Service
      um servio de banco de dados gerenciado para uso de banco de dados que utilizam SQL como linguagem de consulta.
     Ele permite que voc crie bancos de dados na nuvem que so gerenciados pela AWS
     Postgres
     MySQL
     MariaDB
     Oracle
     Microsoft SQL Server
     Aurora (Banco de dados proprietrio da AWS)

Vantagens de usar RDS em comparao com implantar um banco de dados em EC2:

     RDS  um servio gerenciado:
     Provisionamento automatizado, patching do sistema operacional
     Backups contnuos e restaurao para um timestamp especfico (Ponto na Restaurao de Tempo)!
     Painis de monitoramento
     Rplicas de leitura para melhorar o desempenho de leitura
     Configurao de Mltiplas Zonas de Disponibilidade para DR (Recuperao de Desastres)
     Janelas de manuteno para atualizaes
     Capacidade de escalonamento (vertical e horizontal)
     Armazenamento suportado pelo EBS (gp2 ou io1)
     MAS voc no pode SSH em suas instncias




Amazon Aurora

     Aurora  uma tecnologia proprietria da AWS (no open source)
     PostgreSQL e MySQL so ambos suportados como banco de dados do Aurora
     Aurora  "otimizado para nuvem da AWS" e alega melhoria de desempenho de 5x sobre MySQL no RDS, mais de 3x o desempenho do Postgres no RDS
     O armazenamento do Aurora cresce automaticamente em incrementos de 10GB, at 128 TB
     Aurora custa mais do que RDS (20% a mais) - mas  mais eficiente
     No est includo no nvel gratuito



Amazon Aurora Serverless

     Instantiao automatizada do banco de dados e dimensionamento automtico com base no uso real
     PostgreSQL e MySQL so ambos suportados como banco de dados do Aurora Serverless
     No h necessidade de planejamento de capacidade
     Menor sobrecarga de gerenciamento
     Pagamento por segundo, pode ser mais econmico
     Casos de uso: bom para cargas de trabalho pouco frequentes, intermitentes ou imprevisveis...




Voc pode criar snapshots do seu RDS, assim voc pode aumentar a potencia dele, migrar para outra regio, etc....
