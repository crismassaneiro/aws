Amazon EC2

    EC2 é um dos serviços mais populares da AWS

    EC2 = Elastic Compute Cloud = infrastructure as a service

    Em resumo, permite:
        Criar maquinas virtuais (EC2)
        Guardar dados em drivers vitural (EBS)
        Distribuir carregamentos entre maquinas (ELB)
        Dimencionar serviços usando o auto scaling group (ASG)    

    Conhecer sobre o EC2 é fundamental para entender como a Cloud Funciona  

    EC2 tamanho e configuração

    Você possui três opção de sistemas operacionas para criar sua maquina:

    Linux
    Windows
    Mac OS

    Você precisa selecionar o processador da sua máquina (CPU)
    Selecionar a quantidade de memória que você irá precisar (RAM)
    Quanto de armazenamento essa máquina terá:
        network-attached EBS & EFS
        hardware(EC2 instance Store)

    Network card: speed of the card. Pulbic IP address

    Firewall rules: securyt group

    Bootstrap script (configurado no primeiro lançamento da sua máquina): EC2 User Data

EC2 User data

    São comandos que você pode criar através de um scrip, para que, quando sua máquina for startar pela primeira vez ela siga.
        Tipos de comandos:
        Instalar atualizações
        Instalar softwares
        Fazer o download de arquivos da internet
        Tudo o que você quiser....

    Existem várias configurações máquinas, então você sempre terá alguma máquina que conseguirá rodar seus códigos.

    Como lançar uma instância   EC2

    Vá em em EC2, clique em instances, em lanraja: Launch Instances

    Name and tags 
        Crie uma taga para essa instancia

    Application and OS Images (Amazon Machine Image) 
        Selecione o sistema operacional da sun máquina

    Instance type
        Selecione sua instancia

    Network settings
        Deixe tudo padrão

    Configure storage
        Tamanho do armazenamento da sua máquina(deixe padrão)

    Advanced details 
        Deixe padrõa
        Mas se atende que no final, existe um campo em branco, lá você pode colocar um código que só será executado quando sua máquina for lançada. Isso serve para testar ela

                systemctl enable httpd

Security Grups 

    São a base da segurança da AWS
    São responsáveis por controlar o que pode sair e o que pode entrar nas instancias Ec2
    Só possuem regras de permissão
    Eles são a proteção dos nossos Ec2
    Eles regulam:

        Acesso a portas

        IPs ranges IPv4 E IPv6

        Controlam a entrada

        Controlam a saída

    Importante saber:

    Você pode usar um Security Group para várias instâncias
    Uma instancia pode ter vários Secutiry Groups
    Security Grups funcionam por região/VPC, se você mudar, precisa criar outro Security Group para região
    Security Groups não estão dentro do EC2, eles funcionam a parte 
    É muito recomendável manter um Security Group separado para SSH access
    Se a sua aplicação não ficar acessível (time out), por causa de um erro no seu Security Group
    Se aparecer o erro CONEXÃO RECUSA, significa que é um arro na aplicaçaõ ou ela não foi lançada.
    Todo trafico inboud é bloqueado por padrão
    Todo trafico outbound é autorizado por padrão

    Portas que devemos saber para a prova:

    22 = SSH(Secure Shell) - log into a Linux instance
    21 = FTP(File Transfer Protocol - upload files into a file share)
    22 = SFTP(Secure File Transfer Protocol) - upload files using SSH
    80 = HTTPS - access ansecured websites
    443 = HTTPS - access secured websites
    3389 = RDP(Remote Desktop Protocol) - log into a Windows instance

Sobre o SSH:



    Existe duas formas de se conectar com sua instância SSH

    Uma delas é através do seu Desktop, você precisará da sua chave.pem que você faz o download quando criar uma instancia Ec2.

    Outras forma é você se conectar diratamente pelo console, indo em Intances - Ec2: selecione a opção conect

    Nunca salve suas chaves access key e security key dentro da sua instancia ec2
    IAM

    Para que suas intancias Ec2 tenhama cesso a sua chaves, você precisa adicionar IAM Role em sua instancia EC2. 

    Para fazer isso vá até sua intancia, selecione ele, clique em actions, security e Modify IAM Role fazendo isso, você poderá escolher uma iam Role ou criar uma.

Opções de compra de EC2

    Instâncias sob demanda – carga de trabalho curta, preços previsíveis, pagamento por segundo
    • Reservado (1 e 3 anos)
        • Instâncias reservadas – cargas de trabalho longas
        • Instâncias reservadas conversíveis – cargas de trabalho longas com instâncias flexíveis
    • Planos de Poupança (1 e 3 anos) – compromisso com uma quantidade de uso, longa carga de trabalho
    • Instâncias Spot – cargas de trabalho curtas, baratas, podem perder instâncias (menos confiáveis)
    • Hosts Dedicados – reserve um servidor físico inteiro, controle o posicionamento da instância
    • Instâncias dedicadas – nenhum outro cliente compartilhará seu hardware
    • Reservas de Capacidade – reserva de capacidade em uma AZ específica por qualquer duração

EC2 sob demanda

    Pague pelo que você usa:
    Linux ou Windows – cobrança por segundo, após o primeiro minuto
    Todos os outros sistemas operacionais - faturamento por hora
    Tem o custo mais alto, mas sem pagamento antecipado
    Nenhum compromisso de longo prazo
    Recomendado para cargas de trabalho ininterruptas e de curto prazo, onde
    você não pode prever como o aplicativo se comportará

Instâncias reservadas do EC2

    • Até 72% de desconto em comparação com On-demand
    • Você reserva atributos de instância específicos (tipo de instância, região, locação, sistema operacional)
    • Período de Reserva – 1 ano (+desconto) ou 3 anos (+++desconto)
    • Opções de pagamento – sem adiantamento (+), adiantamento parcial (++), tudo adiantado (+++)
    • Escopo da Instância Reservada – Regional ou Zonal (capacidade de reserva em uma AZ)
    • Recomendado para aplicações de uso em estado estacionário (pense em banco de dados)
    • Você pode comprar e vender no Mercado de Instâncias Reservadas
    • Instância reservada conversível
        • Pode alterar o tipo de instância do EC2, a família de instâncias, o sistema operacional, o escopo e a locação
        • Até 66% de desconto

Planos de poupança EC2

    • Obtenha um desconto com base no uso de longo prazo (até 72% – o mesmo que RIs)
    • Comprometa-se com um determinado tipo de uso (US$ 10/hora por 1 ou 3 anos)
    • O uso além dos EC2 Savings Plans é cobrado pelo preço sob demanda
    • Bloqueado para uma família de instâncias e região da AWS específicas (por exemplo, M5 em us-east-1)
    • Flexível em:
        • Tamanho da instância (por exemplo, m5.xlarge, m5.2xlarge)
        • SO (por exemplo, Linux, Windows)
        • Locação (Host, Dedicado, Padrão)

Instâncias spot EC2

    • Pode obter um desconto de até 90% em comparação com On-demand
    • Instâncias que você pode “perder” a qualquer momento se seu preço máximo for menor que o
    preço à vista atual
    • As instâncias MAIS econômicas da AWS
    • Útil para cargas de trabalho resilientes a falhas
    • Trabalhos em lote
    • Análise de dados
    • Processamento de imagem
    • Quaisquer cargas de trabalho distribuídas
    • Cargas de trabalho com horários de início e término flexíveis
    • Não é adequado para trabalhos ou bancos de dados críticos

Hosts dedicados EC2

    • Um servidor físico com capacidade de instância EC2 totalmente dedicada ao seu uso
    • Permite que você atenda aos requisitos de conformidade e use suas licenças de software existentes vinculadas ao servidor (licenças de software por soquete, por núcleo e pe-VM)
    • Opções de compra:
    • Sob demanda – pagamento por segundo para Host Dedicado ativo
    • Reservado - 1 ou 3 anos (sem adiantamento, adiantamento parcial, adiantamento total)
    • A opção mais cara
    • Útil para software com modelo de licenciamento complicado (BYOL – Bring Your Own License)
    • Ou para empresas que têm fortes necessidades regulatórias ou de conformidade

Instâncias dedicadas do EC2 

    • As instâncias são executadas em hardware que Dedicado à você
    • Pode compartilhar hardware com outras instâncias na mesma conta
    • Nenhum controle sobre o posicionamento das instâncias
    (pode mover hardware após Stop/Start)



Reservas de capacidade EC2

    • Reservar capacidade de instâncias sob demanda em uma AZ específica para qualquer
    duração
    • Você sempre terá acesso à capacidade do EC2 quando precisar
    • Sem compromisso de tempo (criar/cancelar a qualquer momento), sem descontos de faturamento
    • Combine com Instâncias Reservadas Regionais e Planos de Poupança para se beneficiar
    de descontos de faturamento
    • Você será cobrado de acordo com a taxa sob demanda, independentemente de executar instâncias ou não
    • Adequado para cargas de trabalho ininterruptas e de curto prazo que precisam estar em um AZ específica
