Amazon EC2

        Criar maquinas virtuais (EC2)

    Importante saber:

    Voc pode usar um Security Group para vrias instncias
    Uma instancia pode ter vrios Secutiry Groups
    Security Grups funcionam por regio/VPC, se voc mudar, precisa criar outro Security Group para regio
    Security Groups no esto dentro do EC2, eles funcionam a parte 
     muito recomendvel manter um Security Group separado para SSH access
    Se a sua aplicao no ficar acessvel (time out), por causa de um erro no seu Security Group
    Se aparecer o erro CONEXO RECUSA, significa que  um arro na aplicaa ou ela no foi lanada.
    Todo trafico inboud  bloqueado por padro
    Todo trafico outbound  autorizado por padro

    Portas que devemos saber para a prova:

    22 = SSH(Secure Shell) - log into a Linux instance
    21 = FTP(File Transfer Protocol - upload files into a file share)
    22 = SFTP(Secure File Transfer Protocol) - upload files using SSH
    80 = HTTPS - access ansecured websites
    443 = HTTPS - access secured websites
    3389 = RDP(Remote Desktop Protocol) - log into a Windows instance

Opes de compra de EC2

    Instncias sob demanda  carga de trabalho curta, preos previsveis, pagamento por segundo
     Reservado (1 e 3 anos)
         Instncias reservadas  cargas de trabalho longas
         Instncias reservadas conversveis  cargas de trabalho longas com instncias flexveis
     Planos de Poupana (1 e 3 anos)  compromisso com uma quantidade de uso, longa carga de trabalho
     Instncias Spot  cargas de trabalho curtas, baratas, podem perder instncias (menos confiveis)
     Hosts Dedicados  reserve um servidor fsico inteiro, controle o posicionamento da instncia
     Instncias dedicadas  nenhum outro cliente compartilhar seu hardware
     Reservas de Capacidade  reserva de capacidade em uma AZ especfica por qualquer durao

EC2 sob demanda

    Pague pelo que voc usa:
    Linux ou Windows  cobrana por segundo, aps o primeiro minuto
    Todos os outros sistemas operacionais - faturamento por hora
    Tem o custo mais alto, mas sem pagamento antecipado
    Nenhum compromisso de longo prazo
    Recomendado para cargas de trabalho ininterruptas e de curto prazo, onde
    voc no pode prever como o aplicativo se comportar

Instncias reservadas do EC2

     At 72% de desconto em comparao com On-demand
     Voc reserva atributos de instncia especficos (tipo de instncia, regio, locao, sistema operacional)
     Perodo de Reserva  1 ano (+desconto) ou 3 anos (+++desconto)
     Opes de pagamento  sem adiantamento (+), adiantamento parcial (++), tudo adiantado (+++)
     Escopo da Instncia Reservada  Regional ou Zonal (capacidade de reserva em uma AZ)
     Recomendado para aplicaes de uso em estado estacionrio (pense em banco de dados)
     Voc pode comprar e vender no Mercado de Instncias Reservadas
     Instncia reservada conversvel
         Pode alterar o tipo de instncia do EC2, a famlia de instncias, o sistema operacional, o escopo e a locao
         At 66% de desconto

Planos de poupana EC2

     Obtenha um desconto com base no uso de longo prazo (at 72%  o mesmo que RIs)
     Comprometa-se com um determinado tipo de uso (US$ 10/hora por 1 ou 3 anos)
     O uso alm dos EC2 Savings Plans  cobrado pelo preo sob demanda
     Bloqueado para uma famlia de instncias e regio da AWS especficas (por exemplo, M5 em us-east-1)
     Flexvel em:
         Tamanho da instncia (por exemplo, m5.xlarge, m5.2xlarge)
         SO (por exemplo, Linux, Windows)
         Locao (Host, Dedicado, Padro)

Instncias spot EC2

     Instncias que voc pode perder a qualquer momento se seu preo mximo for menor que o

Hosts dedicados EC2

     Um servidor fsico com capacidade de instncia EC2 totalmente dedicada ao seu uso
     Permite que voc atenda aos requisitos de conformidade e use suas licenas de software existentes vinculadas ao servidor (licenas de software por soquete, por ncleo e pe-VM)
     Opes de compra:
     Sob demanda  pagamento por segundo para Host Dedicado ativo
     Reservado - 1 ou 3 anos (sem adiantamento, adiantamento parcial, adiantamento total)
     A opo mais cara
     til para software com modelo de licenciamento complicado (BYOL  Bring Your Own License)
     Ou para empresas que tm fortes necessidades regulatrias ou de conformidade

Instncias dedicadas do EC2 

     As instncias so executadas em hardware que Dedicado  voc
     Pode compartilhar hardware com outras instncias na mesma conta
     Nenhum controle sobre o posicionamento das instncias
    (pode mover hardware aps Stop/Start)



Reservas de capacidade EC2

     Reservar capacidade de instncias sob demanda em uma AZ especfica para qualquer
    durao

     Sem compromisso de tempo (criar/cancelar a qualquer momento), sem descontos de faturamento

     Adequado para cargas de trabalho ininterruptas e de curto prazo que precisam estar em um AZ especfica
