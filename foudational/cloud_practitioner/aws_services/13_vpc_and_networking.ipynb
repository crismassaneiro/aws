Endereços IP na AWS

    • IPv4 – Protocolo de Internet versão 4 (4,3 bilhões de endereços)
    • IPv4 público – pode ser usado na Internet
    • A instância EC2 obtém um novo endereço IP público sempre que você a interrompe e inicia (padrão)
    • IPv4 privado – pode ser usado em redes privadas (LAN), como AWS internas
    rede (por exemplo, 192.168.1.1)
    • O IPv4 privado é corrigido para instâncias EC2 mesmo se você as iniciar/interromper
    • Elastic IP – permite anexar um endereço IPv4 público fixo à instância do EC2
    • Observação: tem custo contínuo se não estiver anexado à instância EC2 ou se a instância EC2 for
    parou
    • IPv6 – Protocolo de Internet versão 6 (3,4 × 10!" Endereços)
    • Todo endereço IP é público (sem intervalo privado)
    • Exemplo: 2001:db8:3333:4444:cccc:dddd:eeee:ffff

VPC & Subnets Primer

    • VPC - Virtual Private Cloud: rede privada para implantar seus recursos
    (recurso regional)
    • As sub-redes permitem particionar sua rede dentro da VPC
    (recurso da zona de disponibilidade)
    • Uma sub-rede pública é uma sub-rede que é
    acessível pela internet
    • Uma sub-rede privada é uma sub-rede que não é
    acessível pela internet
    • Definir o acesso à internet e
    entre sub-redes, usamos tabelas de rotas

Internet Gateway & NAT Gateways

    • Internet Gateways ajudam nossa VPC instâncias se conectam à Internet
    • Public Subnets têm uma rota para Internet Gateway.

    • NAT Gateways (gerenciados pela AWS) e NAT instances (autogerenciadas) permitem suas instâncias em suas sub-redes privadas para acessar a internet enquanto permanece privado
    


Network ACL & Security Groups

    NACL (Network ACL)

        • Um firewall que controla o tráfego de e para sub-rede
        • Pode ter regras ALLOW e DENY
        • Estão anexados no nível da sub-rede
        • As regras incluem apenas endereços IP

    • Security Groups

        • Um firewall que controla o tráfego de e para um
        ENI/uma instância EC2
        • Só pode ter regras ALLOW
        • As regras incluem endereços IP e outras informações de segurança
        grupos

VPC Flow Logs

    • Capture informações sobre o tráfego IP que entra em suas interfaces:
    • Registros de fluxo de VPC
    • Logs de fluxo de sub-rede
    • Logs de fluxo da interface de rede elástica
    • Ajuda a monitorar e solucionar problemas de conectividade. Exemplo:
    • Sub-redes para internet
    • Sub-redes para sub-redes
    • Internet para sub-redes
    • Captura informações de rede também de interfaces gerenciadas pela AWS: Elastic Load
    Balanceadores, ElastiCache, RDS, Aurora, etc…
    • Os dados dos logs do VPC Flow podem ir para S3, CloudWatch Logs e Kinesis Data Firehose

VPC Peering

    • Conecte duas VPCs, de forma privada, usando Rede da AWS
    • Faça com que eles se comportem como se fossem na mesma rede
    • Não deve ter CIDR (IP intervalo de endereços)
    • A conexão de peering de VPC não está transitivo (deve ser estabelecido para cada VPC que precisa comunicar um com o outro)



VPC Endpoints

    • Os endpoints permitem que você se conecte aos serviços da AWS usando uma rede privada em vez da rede www pública
     • Isso oferece segurança aprimorada e menor latência para acessar os serviços da AWS
     • Gateway VPC Endpoint: S3 e DynamoDB
     • Interface VPC Endpoint: o resto



AWS PrivateLink (VPC Endpoint Services)

    • A maneira mais segura e escalável de expor um serviço a milhares de VPCs
    • Não requer peering de VPC, gateway de internet, NAT, tabelas de rotas…
    • Requer um balanceador de carga de rede (VPC de serviço) e ENI (VPC do cliente)



Site to Site VPN & Direct Connect

• Site to Site VPN

    • Conecte uma VPN local à AWS
    • A conexão é automaticamente criptografado
    • Passa pela Internet pública

AWS Direct Connect (DX)

    • Establish a physical connection between on-premises and AWS
    • The connection is private, secure and fast
    • Goes over a private network
    • Takes at least a month to establish



Site-to-Site VPN

On-premises: must use a Customer Gateway (CGW)
• AWS: must use a Virtual Private Gateway (VGW)



AWS Client VPN

• Conecte-se do seu computador usando OpenVPN à sua rede privada na AWS e no local
• Permitir que você se conecte às suas instâncias do EC2 por meio de um IP privado (como se você estava na rede VPC privada)
• Passa pela Internet pública

Transit Gateway

For having transitive peering between thousands of VPC and on-premises, hub-and-spoke(star)connection
• One single Gateway to provide this functionality
• Works with Direct Connect Gateway, VPN connections



VPC Summary

    • VPC – Virtual Private Cloud
    • Subnets – Tied to an AZ, network partition of the VPC
    • Internet Gateway – at the VPC level, provide Internet Access
    • NAT Gateway / Instances – give internet access to private subnets
    • NACL – Stateless, subnet rules for inbound and outbound
    • Security Groups – Stateful, operate at the EC2 instance level or ENI
    • VPC Peering – Connect two VPC with non overlapping IP ranges, nontransitive
    • Elastic IP –fixed public IPv4, ongoing cost if not in-use
    • VPC Endpoints – Provide private access to AWS Services within VPC
    • PrivateLink – Privately connect to a service in a 3rd party VPC
    • VPC Flow Logs – network traffic logs
    • Site to Site VPN – VPN over public internet between on-premises DC and AWS
    • Client VPN – OpenVPN connection from your computer into your VPC
    • Direct Connect – direct private connection to AWS
    • Transit Gateway – Connect thousands of VPC and on-premises networks together
