Amazon CloudWatch Metrics

    • O CloudWatch fornece métricas para todos os serviços na AWS.
    • Uma métrica é uma variável para monitorar (CPU Utilization, NetworkIn...).
    • As métricas têm carimbos de data e hora.
    • É possível criar painéis do CloudWatch com métricas.



Métricas importantes

    • Instâncias EC2: utilização de CPU, verificações de status, rede (não RAM)
    • Métricas padrão a cada 5 minutos
    • Opção de monitoramento detalhado ($$$): métricas a cada 1 minuto
    • Volumes EBS: leitura/gravação de disco
    • Buckets S3: BucketSizeBytes, NumberOfObjects, AllRequests
    • Faturamento: cobrança total estimada (somente em us-east-1)
    • Limites de serviço: quanto você está usando uma API de serviço
    • Métricas personalizadas: impulsione suas próprias métricas

Amazon CloudWatch Alarms

    • Alarmes são usados para disparar notificações para qualquer métrica
    • Ações dos alarmes...
    • Auto Scaling: aumentar ou diminuir o contador "desejado" de instâncias EC2
    • Ações EC2: parar, encerrar, reiniciar ou recuperar uma instância EC2
    • Notificações SNS: enviar uma notificação para um tópico SNS
    • Várias opções (amostragem, %, máximo, mínimo, etc...)
    • É possível escolher o período no qual avaliar um alarme
    • Exemplo: criar um alarme de faturamento na métrica de Faturamento do CloudWatch
    • Estados do Alarme: OK, DADOS_INSUFICIENTES, ALARME

Amazon CloudWatch Logs    

    • Os Logs do CloudWatch podem coletar logs de:
    • Elastic Beanstalk: coleta de logs de aplicativos
    • ECS: coleta de logs de contêineres
    • AWS Lambda: coleta de logs de funções
    • CloudTrail com base em filtro
    • Agentes de logs do CloudWatch: em máquinas EC2 ou em servidores locais
    • Route53: Registra consultas DNS
    • Permite monitoramento em tempo real de logs
    • Retenção ajustável de Logs do CloudWatch

CloudWatch Logs for EC2

    • Por padrão, nenhum log da sua instância EC2 será enviado para o CloudWatch.
    • É necessário executar um agente do CloudWatch na EC2 para enviar os arquivos de log desejados.
    • Certifique-se de que as permissões do IAM estejam corretas.
    • O agente de log do CloudWatch pode ser configurado também em servidores locais.

AWS CloudTrail

    • Fornece governança, conformidade e auditoria para sua conta da AWS
    • O CloudTrail está ativado por padrão!
    • Obtenha um histórico de eventos/chamadas de API feitas dentro de sua conta da AWS por:
    • Console
    • SDK
    • CLI
    • Serviços AWS
    • Pode colocar logs do CloudTrail no CloudWatch Logs ou S3
    • Um trilho pode ser aplicado a Todas as Regiões (padrão) ou a uma única Região.
    • Se um recurso for excluído na AWS, investigue o CloudTrail primeiro!
