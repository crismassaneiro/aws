Amazon CloudWatch Metrics

     O CloudWatch fornece mtricas para todos os servios na AWS.
     Uma mtrica  uma varivel para monitorar (CPU Utilization, NetworkIn...).
     As mtricas tm carimbos de data e hora.
      possvel criar painis do CloudWatch com mtricas.



Mtricas importantes

     Instncias EC2: utilizao de CPU, verificaes de status, rede (no RAM)
     Mtricas padro a cada 5 minutos
     Opo de monitoramento detalhado ($$$): mtricas a cada 1 minuto
     Volumes EBS: leitura/gravao de disco
     Buckets S3: BucketSizeBytes, NumberOfObjects, AllRequests
     Faturamento: cobrana total estimada (somente em us-east-1)
     Limites de servio: quanto voc est usando uma API de servio
     Mtricas personalizadas: impulsione suas prprias mtricas

Amazon CloudWatch Alarms

     Alarmes so usados para disparar notificaes para qualquer mtrica
     Aes dos alarmes...
     Auto Scaling: aumentar ou diminuir o contador "desejado" de instncias EC2
     Aes EC2: parar, encerrar, reiniciar ou recuperar uma instncia EC2
     Notificaes SNS: enviar uma notificao para um tpico SNS
     Vrias opes (amostragem, %, mximo, mnimo, etc...)
      possvel escolher o perodo no qual avaliar um alarme
     Exemplo: criar um alarme de faturamento na mtrica de Faturamento do CloudWatch
     Estados do Alarme: OK, DADOS_INSUFICIENTES, ALARME

Amazon CloudWatch Logs    

     Os Logs do CloudWatch podem coletar logs de:
     Elastic Beanstalk: coleta de logs de aplicativos
     ECS: coleta de logs de contineres
     AWS Lambda: coleta de logs de funes
     CloudTrail com base em filtro
     Agentes de logs do CloudWatch: em mquinas EC2 ou em servidores locais
     Route53: Registra consultas DNS
     Permite monitoramento em tempo real de logs
     Reteno ajustvel de Logs do CloudWatch

CloudWatch Logs for EC2

     Por padro, nenhum log da sua instncia EC2 ser enviado para o CloudWatch.
      necessrio executar um agente do CloudWatch na EC2 para enviar os arquivos de log desejados.
     Certifique-se de que as permisses do IAM estejam corretas.
     O agente de log do CloudWatch pode ser configurado tambm em servidores locais.

AWS CloudTrail

     Fornece governana, conformidade e auditoria para sua conta da AWS
     O CloudTrail est ativado por padro!
     Obtenha um histrico de eventos/chamadas de API feitas dentro de sua conta da AWS por:
     Console
     SDK
     CLI
     Servios AWS
     Pode colocar logs do CloudTrail no CloudWatch Logs ou S3
     Um trilho pode ser aplicado a Todas as Regies (padro) ou a uma nica Regio.
     Se um recurso for excludo na AWS, investigue o CloudTrail primeiro!
