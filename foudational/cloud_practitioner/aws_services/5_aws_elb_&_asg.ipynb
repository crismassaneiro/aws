Scalability & High Availability

PENSE EM AVIÕES NA ANALOGIA....

Scalabilidade significa que sua aplicação/sistema, pode lidar com cargas maiores se adaptando

Existe 2 tipos de escalabilidade:
    Vertical
    Horizontal
    
Vertical Scaling

Significa crescer o poder de um intãncia
Pensa assim, você tem 1 EC2 podre e então para aumentar seu poder computacional você faz um upgrade para um EC2 superior

Mas querendo ou não, você chegará em um limite de instância, porque, por mais que sua máquina seja a mais forte da AWS, em algum momento ela chegará no limite, dependendo do seu cenário. Fora questões de custo. 
Muito comum em data bases, isso ocorre pq o tempo de processamento aumenta conforme a base cresce, então você acaba precisando de mais poder computacional para rodar o código

Horizontal Scaling
Você possui um EC2 e precisa aumentar o poder computacional, então você cria outra intância EC2, identica, se precisar, criar outra e outra, até suprir sua necessidade.
Significa crescer o número de intâncias

High Availability

Anda de mãos dadas com escalonamento horizontal
Significa também que suas aplicação está rodando pelo menos por 2 zonas
Significa que, se o servidor Nova York cair, tem o São Francisco, isso signfica grande disponibilidade, pq sempre estará rodando sua aplicação, mesmo que ocorra uma tragédia em algum data center.

O que é Load Balancing

Balanceadores de carga são servidores que encaminham o tráfego da Internet para vários servidores (instâncias EC2) downstream



Por que usar um load Balancer? 

    • Distribua a carga entre diversas instâncias downstream 
    • Exponha um único ponto de acesso (DNS) para seu aplicativo 
    • Lide perfeitamente com falhas de instâncias downstream 
    • Faça verificações de integridade regulares em suas instâncias 
    • Forneça terminação SSL (HTTPS) para seus sites 
    • Alta disponibilidade em todas as zonas


Por que usar um ELB (Elastic Load Balancer)

    • Um ELB (Elastic Load Balancer) é um balanceador de carga gerenciado
    • A AWS garante que estará funcionando
    • A AWS cuida de atualizações, manutenção, alta disponibilidade
    • A AWS fornece apenas alguns ajustes de configuração
    • Custa menos configurar seu próprio balanceador de carga, mas exigirá muito mais esforço de sua parte (manutenção, integrações)
    • 4 tipos de balanceadores de carga oferecidos pela AWS:
    • Application Load Balancer (apenas HTTP / HTTPS) - Layer 7
    • Network Load Balancer (ultra-alto desempenho, permite TCP) - Layer 4
    • Gateway Load Balancer - Layer 3
    • Classic Load Balancer (aposentado em 2023) - Layers 4 e 7



Para criar um load balancer:

    Vá até load balancer
    Criar load balancer
    Selecione o tipo que você deseja, vamos de ALB (siglas de Aplication Load Balancer)
    De um nome
    Selecione o security group que você deseja
    Na parte de Listeners and routing, é onde você selecione as instâncias EC2 que você deseja por no load balancer
    Caso já tenha configurado, basta selecionar o grupo, caso contrário, crie (lembre-se de confirar corretamente)
    Clique em Next
    Selecione as intâncias que você deseja
    Seja feliz, você criou um load balancer
    Para testar, abra o link em Load Balancers - DNS Name e fique dando f5, verá que o IP de instância fica mudando, isso pq ele está distribuindo o tráfego entre diferentes instâcias

Auto Scaling Group

O que é um Grupo de Auto Scaling?
    • Na vida real, a carga nos seus sites e aplicativos pode mudar.
    • Na nuvem, você pode criar e se livrar de servidores muito rapidamente.
    • O objetivo de um Grupo de Auto Scaling (ASG) é:
    • Escalar para fora (adicionar instâncias EC2) para corresponder a uma carga aumentada.
    • Escalar para dentro (remover instâncias EC2) para corresponder a uma carga diminuída.
    • Garantir que tenhamos um número mínimo e máximo de máquinas em execução.
    • Registrar automaticamente novas instâncias em um balanceador de carga.
    • Substituir instâncias não saudáveis.
    • Economia de custos: operar apenas com capacidade ótima (princípio da nuvem).






Para criar um Auto Scaling Groups:

    Vá até Auto Scaling
    Crie um grupo
    Selecione o número de instâncias que esse grupo deverá ter
    e é basicamente isso, bem resumidamente
    obs: se você configurar para ter 2 intancias e um intancia for deletada ou der problema, o Auto Scaling Group criado um nova automaticamente!

Grupos de Auto Scaling - Estratégias de Dimensionamento

    Dimensionamento Manual: Atualize o tamanho de um ASG manualmente.
    Dimensionamento Dinâmico: Responda à demanda em mudança.
    Dimensionamento Simples / Por Etapas:
      Quando um alarme do CloudWatch é acionado (por exemplo, CPU > 70%), então adicione 2 unidades.
       Quando um alarme do CloudWatch é acionado (por exemplo, CPU < 30%), então remova 1 unidade.
    Dimensionamento por Rastreamento de Alvo:
        Exemplo: Eu quero que a CPU média do ASG fique em torno de 40%.
    Dimensionamento Agendado:
        Antecipe um dimensionamento com base em padrões de uso conhecidos.
        Exemplo: aumente a capacidade mínima para 10 às 17h às sextas-feiras.

Dimensionamento Preditivo:
    Usa Aprendizado de Máquina para prever o tráfego futuro com antecedência.
    Provisiona automaticamente o número correto de instâncias EC2 com antecedência.
    Útil quando sua carga tem padrões de tempo previsíveis.



ELB & ASG - Resumo

Alta Disponibilidade vs Escalabilidade (vertical e horizontal) vs Elasticidade vs Agilidade na Nuvem
Balanceadores de Carga Elásticos (ELB)
    Distribuem o tráfego entre instâncias EC2 no backend, podem ser Multi-AZ
    Suportam verificações de integridade
    4 tipos: Clássico (antigo), Application (HTTP - L7), Network (TCP - L4), Gateway (L3)
Grupos de Auto Scaling (ASG)
    Implementam Elasticidade para sua aplicação, em várias zonas de disponibilidade
    Dimensionam instâncias EC2 com base na demanda do seu sistema, substituem instâncias não saudáveis
