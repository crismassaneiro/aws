Scalability & High Availability

PENSE EM AVIES NA ANALOGIA....

Scalabilidade significa que sua aplicao/sistema, pode lidar com cargas maiores se adaptando

Existe 2 tipos de escalabilidade:
    Vertical
    Horizontal
    
Vertical Scaling

Significa crescer o poder de um intncia
Pensa assim, voc tem 1 EC2 podre e ento para aumentar seu poder computacional voc faz um upgrade para um EC2 superior

Mas querendo ou no, voc chegar em um limite de instncia, porque, por mais que sua mquina seja a mais forte da AWS, em algum momento ela chegar no limite, dependendo do seu cenrio. Fora questes de custo. 
Muito comum em data bases, isso ocorre pq o tempo de processamento aumenta conforme a base cresce, ento voc acaba precisando de mais poder computacional para rodar o cdigo

Horizontal Scaling
Voc possui um EC2 e precisa aumentar o poder computacional, ento voc cria outra intncia EC2, identica, se precisar, criar outra e outra, at suprir sua necessidade.
Significa crescer o nmero de intncias

High Availability

Anda de mos dadas com escalonamento horizontal
Significa tambm que suas aplicao est rodando pelo menos por 2 zonas
Significa que, se o servidor Nova York cair, tem o So Francisco, isso signfica grande disponibilidade, pq sempre estar rodando sua aplicao, mesmo que ocorra uma tragdia em algum data center.

O que  Load Balancing

Balanceadores de carga so servidores que encaminham o trfego da Internet para vrios servidores (instncias EC2) downstream



Por que usar um load Balancer? 

     Distribua a carga entre diversas instncias downstream 
     Exponha um nico ponto de acesso (DNS) para seu aplicativo 
     Lide perfeitamente com falhas de instncias downstream 
     Faa verificaes de integridade regulares em suas instncias 
     Fornea terminao SSL (HTTPS) para seus sites 
     Alta disponibilidade em todas as zonas


Por que usar um ELB (Elastic Load Balancer)

     Um ELB (Elastic Load Balancer)  um balanceador de carga gerenciado
     A AWS garante que estar funcionando
     A AWS cuida de atualizaes, manuteno, alta disponibilidade
     A AWS fornece apenas alguns ajustes de configurao
     Custa menos configurar seu prprio balanceador de carga, mas exigir muito mais esforo de sua parte (manuteno, integraes)
     4 tipos de balanceadores de carga oferecidos pela AWS:
     Application Load Balancer (apenas HTTP / HTTPS) - Layer 7
     Network Load Balancer (ultra-alto desempenho, permite TCP) - Layer 4
     Gateway Load Balancer - Layer 3
     Classic Load Balancer (aposentado em 2023) - Layers 4 e 7



Para criar um load balancer:

    V at load balancer
    Criar load balancer
    Selecione o tipo que voc deseja, vamos de ALB (siglas de Aplication Load Balancer)
    De um nome
    Selecione o security group que voc deseja
    Na parte de Listeners and routing,  onde voc selecione as instncias EC2 que voc deseja por no load balancer
    Caso j tenha configurado, basta selecionar o grupo, caso contrrio, crie (lembre-se de confirar corretamente)
    Clique em Next
    Selecione as intncias que voc deseja
    Seja feliz, voc criou um load balancer
    Para testar, abra o link em Load Balancers - DNS Name e fique dando f5, ver que o IP de instncia fica mudando, isso pq ele est distribuindo o trfego entre diferentes instcias

Auto Scaling Group

O que  um Grupo de Auto Scaling?
     Na vida real, a carga nos seus sites e aplicativos pode mudar.
     Na nuvem, voc pode criar e se livrar de servidores muito rapidamente.
     O objetivo de um Grupo de Auto Scaling (ASG) :
     Escalar para fora (adicionar instncias EC2) para corresponder a uma carga aumentada.
     Escalar para dentro (remover instncias EC2) para corresponder a uma carga diminuda.
     Garantir que tenhamos um nmero mnimo e mximo de mquinas em execuo.
     Registrar automaticamente novas instncias em um balanceador de carga.
     Substituir instncias no saudveis.
     Economia de custos: operar apenas com capacidade tima (princpio da nuvem).






Para criar um Auto Scaling Groups:

    V at Auto Scaling
    Crie um grupo
    Selecione o nmero de instncias que esse grupo dever ter
    e  basicamente isso, bem resumidamente
    obs: se voc configurar para ter 2 intancias e um intancia for deletada ou der problema, o Auto Scaling Group criado um nova automaticamente!

Grupos de Auto Scaling - Estratgias de Dimensionamento

    Dimensionamento Manual: Atualize o tamanho de um ASG manualmente.
    Dimensionamento Dinmico: Responda  demanda em mudana.
    Dimensionamento Simples / Por Etapas:
      Quando um alarme do CloudWatch  acionado (por exemplo, CPU > 70%), ento adicione 2 unidades.
       Quando um alarme do CloudWatch  acionado (por exemplo, CPU < 30%), ento remova 1 unidade.
    Dimensionamento por Rastreamento de Alvo:
        Exemplo: Eu quero que a CPU mdia do ASG fique em torno de 40%.
    Dimensionamento Agendado:
        Antecipe um dimensionamento com base em padres de uso conhecidos.
        Exemplo: aumente a capacidade mnima para 10 s 17h s sextas-feiras.

Dimensionamento Preditivo:
    Usa Aprendizado de Mquina para prever o trfego futuro com antecedncia.
    Provisiona automaticamente o nmero correto de instncias EC2 com antecedncia.
    til quando sua carga tem padres de tempo previsveis.



ELB & ASG - Resumo

Alta Disponibilidade vs Escalabilidade (vertical e horizontal) vs Elasticidade vs Agilidade na Nuvem
Balanceadores de Carga Elsticos (ELB)
    Distribuem o trfego entre instncias EC2 no backend, podem ser Multi-AZ
    Suportam verificaes de integridade

Grupos de Auto Scaling (ASG)
    Implementam Elasticidade para sua aplicao, em vrias zonas de disponibilidade
    Dimensionam instncias EC2 com base na demanda do seu sistema, substituem instncias no saudveis
