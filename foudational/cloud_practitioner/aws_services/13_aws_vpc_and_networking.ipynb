Endereos IP na AWS

     IPv4  Protocolo de Internet verso 4 (4,3 bilhes de endereos)
     IPv4 pblico  pode ser usado na Internet
     A instncia EC2 obtm um novo endereo IP pblico sempre que voc a interrompe e inicia (padro)
     IPv4 privado  pode ser usado em redes privadas (LAN), como AWS internas
    rede (por exemplo, 192.168.1.1)
     O IPv4 privado  corrigido para instncias EC2 mesmo se voc as iniciar/interromper
     Elastic IP  permite anexar um endereo IPv4 pblico fixo  instncia do EC2
     Observao: tem custo contnuo se no estiver anexado  instncia EC2 ou se a instncia EC2 for
    parou
     IPv6  Protocolo de Internet verso 6 (3,4  10!" Endereos)
     Todo endereo IP  pblico (sem intervalo privado)
     Exemplo: 2001:db8:3333:4444:cccc:dddd:eeee:ffff

    acessvel pela internet
     Uma sub-rede privada  uma sub-rede que no 
    acessvel pela internet

Internet Gateway & NAT Gateways

     Internet Gateways ajudam nossa VPC instncias se conectam  Internet

     NAT Gateways (gerenciados pela AWS) e NAT instances (autogerenciadas) permitem suas instncias em suas sub-redes privadas para acessar a internet enquanto permanece privado
    


Network ACL & Security Groups

    NACL (Network ACL)

         Um firewall que controla o trfego de e para sub-rede
         Pode ter regras ALLOW e DENY
         Esto anexados no nvel da sub-rede
         As regras incluem apenas endereos IP

     Security Groups

         Um firewall que controla o trfego de e para um
        ENI/uma instncia EC2
         S pode ter regras ALLOW
         As regras incluem endereos IP e outras informaes de segurana
        grupos

VPC Flow Logs

     Capture informaes sobre o trfego IP que entra em suas interfaces:
     Registros de fluxo de VPC
     Logs de fluxo de sub-rede
     Logs de fluxo da interface de rede elstica
     Ajuda a monitorar e solucionar problemas de conectividade. Exemplo:
     Sub-redes para internet
     Sub-redes para sub-redes
     Internet para sub-redes
     Captura informaes de rede tambm de interfaces gerenciadas pela AWS: Elastic Load
    Balanceadores, ElastiCache, RDS, Aurora, etc
     Os dados dos logs do VPC Flow podem ir para S3, CloudWatch Logs e Kinesis Data Firehose

VPC Peering

     Conecte duas VPCs, de forma privada, usando Rede da AWS
     Faa com que eles se comportem como se fossem na mesma rede
     No deve ter CIDR (IP intervalo de endereos)
     A conexo de peering de VPC no est transitivo (deve ser estabelecido para cada VPC que precisa comunicar um com o outro)



VPC Endpoints

     Os endpoints permitem que voc se conecte aos servios da AWS usando uma rede privada em vez da rede www pblica
      Isso oferece segurana aprimorada e menor latncia para acessar os servios da AWS
      Gateway VPC Endpoint: S3 e DynamoDB
      Interface VPC Endpoint: o resto



AWS PrivateLink (VPC Endpoint Services)

     A maneira mais segura e escalvel de expor um servio a milhares de VPCs
     No requer peering de VPC, gateway de internet, NAT, tabelas de rotas
     Requer um balanceador de carga de rede (VPC de servio) e ENI (VPC do cliente)



Site to Site VPN & Direct Connect

 Site to Site VPN

     Conecte uma VPN local  AWS
     A conexo  automaticamente criptografado
     Passa pela Internet pblica

AWS Direct Connect (DX)

     Establish a physical connection between on-premises and AWS
     The connection is private, secure and fast
     Goes over a private network
     Takes at least a month to establish



Site-to-Site VPN

On-premises: must use a Customer Gateway (CGW)
 AWS: must use a Virtual Private Gateway (VGW)



AWS Client VPN

 Conecte-se do seu computador usando OpenVPN  sua rede privada na AWS e no local
 Permitir que voc se conecte s suas instncias do EC2 por meio de um IP privado (como se voc estava na rede VPC privada)
 Passa pela Internet pblica

Transit Gateway

For having transitive peering between thousands of VPC and on-premises, hub-and-spoke(star)connection
 One single Gateway to provide this functionality
 Works with Direct Connect Gateway, VPN connections



VPC Summary

     VPC  Virtual Private Cloud
     Subnets  Tied to an AZ, network partition of the VPC
     Internet Gateway  at the VPC level, provide Internet Access
     NAT Gateway / Instances  give internet access to private subnets
     NACL  Stateless, subnet rules for inbound and outbound
     Security Groups  Stateful, operate at the EC2 instance level or ENI
     VPC Peering  Connect two VPC with non overlapping IP ranges, nontransitive
     Elastic IP fixed public IPv4, ongoing cost if not in-use
     VPC Endpoints  Provide private access to AWS Services within VPC
     PrivateLink  Privately connect to a service in a 3rd party VPC
     VPC Flow Logs  network traffic logs
     Site to Site VPN  VPN over public internet between on-premises DC and AWS
     Client VPN  OpenVPN connection from your computer into your VPC
     Direct Connect  direct private connection to AWS
     Transit Gateway  Connect thousands of VPC and on-premises networks together
