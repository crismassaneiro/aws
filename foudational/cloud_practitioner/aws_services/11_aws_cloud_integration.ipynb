Section Introduction

    • Esses serviços podem escalar independentemente de nossa aplicação!

Amazon SQS – Standard Queue

    • O mais antigo serviço da AWS (com mais de 10 anos)
    • Serviço totalmente gerenciado (~sem servidor), usado para desacoplar aplicações
    • Escala de 1 mensagem por segundo para dezenas de milhares por segundo
    • Retenção padrão de mensagens: 4 dias, máximo de 14 dias
    • Sem limite para a quantidade de mensagens na fila
    • Mensagens são deletadas após serem lidas pelos consumidores
    • Baixa latência (<10 ms na publicação e recebimento)
    • Consumidores compartilham o trabalho de ler mensagens e escalam horizontalmente

SQS to decouple between application tiers



Amazon SQS – FIFO Queue

    • FIFO = First In First Out (ordenação das mensagens na fila)



    • As mensagens são processadas na ordem pelo consumidor.

Amazon Kinesis

    • Para o exame: Kinesis = streaming de big data em tempo real
    • Serviço gerenciado para coletar, processar e analisar dados de streaming em tempo real em qualquer escala.
    • Muito detalhado para o exame de Cloud Practitioner, mas bom saber:
    • Kinesis Data Streams: streaming de baixa latência para ingestão de dados em escala de centenas de milhares de fontes.
    • Kinesis Data Firehose: carrega streams no S3, Redshift, ElasticSearch, etc.
    • Kinesis Data Analytics: realiza análises em tempo real em streams usando SQL.
    • Kinesis Video Streams: monitora streams de vídeo em tempo real para análise ou aprendizado de máquina.

 Amazon SNS

    • Você consegue criar disparo de emails com isso kkkk
    • Os "editores de eventos" enviam mensagens apenas para um tópico SNS.
    • Podemos ter quantos "assinantes de eventos" quisermos ouvindo as notificações do tópico SNS.
    • Cada assinante do tópico receberá todas as mensagens.
    • Até 12.500.000 assinaturas por tópico, limite de 100.000 tópicos.



 Amazon MQ

    • SQS, SNS são serviços "nativos da nuvem": protocolos proprietários da AWS.
    • Aplicações tradicionais em execução localmente podem usar protocolos abertos como: MQTT, AMQP, STOMP, Openwire, WSS.
    • Ao migrar para a nuvem, em vez de reestruturar a aplicação para usar SQS e SNS, podemos usar o Amazon MQ.
    • O Amazon MQ é um serviço de corretor de mensagens gerenciado para.



    • O Amazon MQ não "escala" tanto quanto o SQS / SNS.
    • O Amazon MQ roda em servidores, pode ser executado em várias Zonas de Disponibilidade (Multi-AZ) com failover.
    • O Amazon MQ possui tanto recursos de fila (~SQS) quanto recursos de tópico (~SNS).

Integration Section – Summary 

    • SQS:
        • Serviço de fila na AWS
        • Múltiplos Produtores, mensagens são mantidas por até 14 dias
        • Múltiplos Consumidores compartilham a leitura e deletam as mensagens quando concluídas
        • Usado para desacoplar aplicações na AWS

    • SNS:
        • Serviço de notificação na AWS
        • Assinantes: Email, Lambda, SQS, HTTP, Mobile...
        • Múltiplos Assinantes, envia todas as mensagens para todos eles
        • Sem retenção de mensagens
        
    • Kinesis:
        • Streaming de dados em tempo real, persistência e análise

    • Amazon MQ:
        • Corretor de mensagens gerenciado para ActiveMQ e RabbitMQ na nuvem (protocolos MQTT, AMQP, etc.).
