Amazon SQS  Standard Queue

     O mais antigo servio da AWS (com mais de 10 anos)
     Servio totalmente gerenciado (~sem servidor), usado para desacoplar aplicaes
     Escala de 1 mensagem por segundo para dezenas de milhares por segundo
     Reteno padro de mensagens: 4 dias, mximo de 14 dias
     Sem limite para a quantidade de mensagens na fila
     Mensagens so deletadas aps serem lidas pelos consumidores
     Baixa latncia (<10 ms na publicao e recebimento)
     Consumidores compartilham o trabalho de ler mensagens e escalam horizontalmente

Amazon Kinesis

     Para o exame: Kinesis = streaming de big data em tempo real
     Servio gerenciado para coletar, processar e analisar dados de streaming em tempo real em qualquer escala.
     Muito detalhado para o exame de Cloud Practitioner, mas bom saber:
     Kinesis Data Streams: streaming de baixa latncia para ingesto de dados em escala de centenas de milhares de fontes.
     Kinesis Data Firehose: carrega streams no S3, Redshift, ElasticSearch, etc.
     Kinesis Data Analytics: realiza anlises em tempo real em streams usando SQL.
     Kinesis Video Streams: monitora streams de vdeo em tempo real para anlise ou aprendizado de mquina.

 Amazon MQ

     SQS, SNS so servios "nativos da nuvem": protocolos proprietrios da AWS.
     Aplicaes tradicionais em execuo localmente podem usar protocolos abertos como: MQTT, AMQP, STOMP, Openwire, WSS.
     Ao migrar para a nuvem, em vez de reestruturar a aplicao para usar SQS e SNS, podemos usar o Amazon MQ.
     O Amazon MQ  um servio de corretor de mensagens gerenciado para.



     O Amazon MQ no "escala" tanto quanto o SQS / SNS.
     O Amazon MQ roda em servidores, pode ser executado em vrias Zonas de Disponibilidade (Multi-AZ) com failover.
     O Amazon MQ possui tanto recursos de fila (~SQS) quanto recursos de tpico (~SNS).

Integration Section  Summary 

     SQS:
         Servio de fila na AWS
         Mltiplos Produtores, mensagens so mantidas por at 14 dias
         Mltiplos Consumidores compartilham a leitura e deletam as mensagens quando concludas
         Usado para desacoplar aplicaes na AWS

     SNS:
         Servio de notificao na AWS
         Assinantes: Email, Lambda, SQS, HTTP, Mobile...
         Mltiplos Assinantes, envia todas as mensagens para todos eles
         Sem reteno de mensagens
        
     Kinesis:
         Streaming de dados em tempo real, persistncia e anlise

     Amazon MQ:
         Corretor de mensagens gerenciado para ActiveMQ e RabbitMQ na nuvem (protocolos MQTT, AMQP, etc.).
