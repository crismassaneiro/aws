EC2 Intance Storage

    Existem diferentes tipos de storages para intancias EC2

EBS = ELastick Block Store

    Ele  uma opo de storage para nossas intancias EC2
    Ele permite que seus dados fiquem salvos, mesmo depois de voce destruir sua instancia EC2
    Eles so vnculados a regies e zonas, ento se voc criou na us-east-1, quer dizer que s nessa zona ela ira aparecer
    Analogia: pense que ele  tipo um pendrive, mas na verso Cloud
    No tier free, temos 30gbs para brincar
    Podemos enexar 1 EBS a vrias instancias de EC2 chama se EBS Multi-Attach

    Quando voc cria uma instancia EC2, automaticamente ser criado um EBS, pq voc precisa configurar um Storage na hora de criar 1 EC2. 

    Quando voc deleta a instancia EC2, automaticamente deletara seu EBS, isso ocorre pq por padro o EBS configura para deletar junto com a intancia EC2, mas voc pode configurar para que, quando o Ec2 for deletado, a intncia EBS permanea.

    Lembrando que 1 EC2 pode ter vrios EBS.

    O EBS precisa ser criado na mesma regio e zona do seu EC2

    Voc pode criar EBSs indo no menu esquerdo e clicando em Elastic Block Store - Volumes

EBS - Snapshot

    Serve basicamente para voc fazer o backup do seu EBS, dessa fomra voc pode tranfereir os dados ou apenas salvalos.

    Quando voc deleta um snapshot, ele ser direcionado para lixeira, nesse processo voc poder restaurar o seu snapshot em at 1 ano. Isso se voc configurar essa opo.

    O snapshot serve tambm para voc fazer a tranferencia de EBS por regies/zonas. Ex: tu pode fazer o snapshot da us-east-1 e levar ele para us-west-1.

EBS Snapshot archive

    Essa verso  75% mais barata que a Snapshot padro
    Porm essa verso mais barata, demora de 24 a 72 horas para restaurar o storage

    AMI = Amazon Machine Image

    O que ?  uma customizao de EC2, onde voc intala seu softaware, configurao, sistema operacional e monitoramento

    AMI  criado de uma regio especifica e pode ser copiado para outras regies

    Voc pode lanar sua intancia EC2 de:

    A public AMI: fornecida pela AWS
    Sua prpria AMI: voc criou e mantem uma AMI
    Por AWS marketplace AMI: um AMI vendido por um terceiro e voc vai l e compra o AMI para poder usar

Como criar um AMI?

    Voc cria uma intncia EC2 e customiza ela
    Para a instancia
    E constroi seu AMI, isso tambm criar EBS snapshots por consequencia....

    Para criar uma AMI, voc clica na instancia EC2 que voc deseja criar a imagem:
    V em actions
    Image and templates
    Create image

    Nisso voc coloca o nome da sua Imagem e colaca a descrio.

    Depois de fazer isso, voc deve ir no local onde ficam as imagens, l estar a imagem que voc copiou da Instncia EC2 -> Amazon Machine Images (AMIs)

    Voc seleciona a AMI que voc deseja lanar 
    configura ela
    e seja feliz

EC2 Intance Store

    Uma opo ao EBS, mas em algumas coisas
    Por que? pq o EBS  bom, mas no  to performtico quanto o Instance Store
    O Instance Store  bora para Buffer / cache / scratj data / temporary content
    Existe risco de perder os dados caso o hardware falhe
    Backups so nossa responsabilidade, no da AWS

    EFS - Elastic File System
    Terceira opo de armazenamento de dados
    Ele pode ser compartilhado em centenas de intancias EC2 ao mesmo tempo
    EFS s funciona com Linux in mult AZ
    Ele  muito abrangente, esclavel e caro a gente paga pelo uso dele
