EC2 Intance Storage

    Existem diferentes tipos de storages para intancias EC2

EBS = ELastick Block Store

    Ele é uma opção de storage para nossas intancias EC2
    Ele permite que seus dados fiquem salvos, mesmo depois de voce destruir sua instancia EC2
    Eles são vínculados a regiões e zonas, então se você criou na us-east-1, quer dizer que só nessa zona ela ira aparecer
    Analogia: pense que ele é tipo um pendrive, mas na versão Cloud
    No tier free, temos 30gbs para brincar
    Podemos enexar 1 EBS a várias instancias de EC2 chama se EBS Multi-Attach

    Quando você cria uma instancia EC2, automaticamente será criado um EBS, pq você precisa configurar um Storage na hora de criar 1 EC2. 

    Quando você deleta a instancia EC2, automaticamente deletara seu EBS, isso ocorre pq por padrão o EBS configura para deletar junto com a intancia EC2, mas você pode configurar para que, quando o Ec2 for deletado, a intância EBS permaneça.

    Lembrando que 1 EC2 pode ter vários EBS.

    O EBS precisa ser criado na mesma região e zona do seu EC2

    Você pode criar EBSs indo no menu esquerdo e clicando em Elastic Block Store - Volumes

EBS - Snapshot

    Serve basicamente para você fazer o backup do seu EBS, dessa fomra você pode tranfereir os dados ou apenas salvalos.

    Quando você deleta um snapshot, ele será direcionado para lixeira, nesse processo você poderá restaurar o seu snapshot em até 1 ano. Isso se você configurar essa opção.

    O snapshot serve também para você fazer a tranferencia de EBS por regiões/zonas. Ex: tu pode fazer o snapshot da us-east-1 e levar ele para us-west-1.

EBS Snapshot archive

    Essa versão é 75% mais barata que a Snapshot padrão
    Porém essa versão mais barata, demora de 24 a 72 horas para restaurar o storage

    AMI = Amazon Machine Image

    O que é? é uma customização de EC2, onde você intala seu softaware, configuração, sistema operacional e monitoramento

    AMI é criado de uma região especifica e pode ser copiado para outras regiões

    Você pode lançar sua intancia EC2 de:

    A public AMI: fornecida pela AWS
    Sua própria AMI: você criou e mantem uma AMI
    Por AWS marketplace AMI: um AMI vendido por um terceiro e você vai lá e compra o AMI para poder usar

Como criar um AMI?

    Você cria uma intância EC2 e customiza ela
    Para a instancia
    E constroi seu AMI, isso também criará EBS snapshots por consequencia....

    Para criar uma AMI, você clica na instancia EC2 que você deseja criar a imagem:
    Vá em actions
    Image and templates
    Create image

    Nisso você coloca o nome da sua Imagem e colaca a descrição.

    Depois de fazer isso, você deve ir no local onde ficam as imagens, lá estará a imagem que você copiou da Instância EC2 -> Amazon Machine Images (AMIs)

    Você seleciona a AMI que você deseja lançar 
    configura ela
    e seja feliz

EC2 Intance Store

    Uma opção ao EBS, mas em algumas coisas
    Por que? pq o EBS é bom, mas não é tão performático quanto o Instance Store
    O Instance Store é bora para Buffer / cache / scratj data / temporary content
    Existe risco de perder os dados caso o hardware falhe
    Backups são nossa responsabilidade, não da AWS

    EFS - Elastic File System
    Terceira opção de armazenamento de dados
    Ele pode ser compartilhado em centenas de intancias EC2 ao mesmo tempo
    EFS só funciona com Linux in mult AZ
    Ele é muito abrangente, esclavel e caro a gente paga pelo uso dele

EFS Infrequent Access (EFS-IA)

    É um EFS para arquivos que não muito acessados
    ELe é 92% mais barto que o EFS padrão
    Se você habilitar, o EFS vai mover automaticamente os arquicos para EFS-IA, com base na ultima vez que foram acessados
    Para fazer isso, você cria uma Lifecycle Policy e diz, ei, em 90 dias eu quero que você mova os arquivos que não estão sendo usados para o EFS-IA, dessa maneira a gente economiza mais né...

RESUMÃO

    EBS:
        É é o storage mais simples do EC2
        Um EC2 pode ter vários EBSs
        Eles são restritos por A-Z
        Podemos usar snapshots para salvar nossos EBSs e tranferelos para outras A-Z

    AMI:
        É uma cópia de um EC2, ex: você configurou um EC2, para não fazer tudo dnv, você salva essa config de EC2 em um AMI, assim você pode lançar sua EC2 configura quantas vezes você quiser, só trocar onome...

    EC2:
        Imagem Builder: automaticamente constroi, testa e distribui AMIs

    EC2 Instance Store:
        É uma opção de armazenamento de altas performance para intancias de EC2
        Perdemos nosso EC2 Instance Store caso a intancia EC2 que ele está alocada seja parada/deletada

    EFS:
        Podemos compartilhar 1 EFS entre centenas de intancias EC2
        É o EBS compartilhado, pq no EBS, ele só pode ficar restrito em uma intancia EC2, mas o EFS pode ser compartilhado entro várias intancias EC2

    EFS-IA:
        É a versão ecnomia do EFS, se você não usa arquivos a muito tempo dentro do seus EFS, você pode criar uma Policy que diz que: em x tempo se um dado não for usado, ele deve ser tranferido para o modo IA, gerando uma economia de até 92%

    Amazon FSx for Windows File Server:
        Exclusivo para arquivos que estão em servidores Windows

    Amazon FSx for Lustre:
        Exclusivo para arquivos que estão em servidores Linux
