Docker

O que  o Docker?

     Docker  uma plataforma de desenvolvimento de software para implantar aplicativos
     Aplicativos so empacotados em contineres que podem ser executados em qualquer sistema       operacional
     Aplicativos funcionam da mesma forma, independentemente de onde so executados
     Qualquer mquina
     Sem problemas de compatibilidade
     Comportamento previsvel
     Menos trabalho
     Mais fcil de manter e implantar
     Funciona com qualquer linguagem, qualquer sistema operacional, qualquer tecnologia
     Escala contineres para cima e para baixo muito rapidamente (segundos)

    

    
Onde as imagens do Docker so armazenadas?

     As imagens do Docker so armazenadas em Repositrios do Docker
     Pblico: Docker Hub https://hub.docker.com/
     Encontre imagens base para muitas tecnologias ou sistemas operacionais:
     Ubuntu
     MySQL
     NodeJS, Java...
     Privado: Amazon ECR (Elastic Container Registry)

Docker versus Mquinas Virtuais

     Docker  "quase" uma tecnologia de virtualizao, mas no exatamente
     Recursos so compartilhados com o host => muitos contineres em um servidor



ECS

    ECS = Elastic Container Service
     Lana contineres Docker na AWS
     Voc precisa provisionar e manter a infraestrutura (as instncias EC2)
     A AWS cuida de iniciar / parar os contineres
     Possui integraes com o Application Load Balancer



Fargate

    Launch Docker containers on AWS
     Voc no provisiona a infraestrutura (sem instncias EC2 para gerenciar) - mais simples!
     Oferta serverless
     A AWS apenas executa os contineres para voc com base na CPU / RAM que voc precisa



ECR

    Elastic Container Registry
     Registro privado do Docker na AWS
     Aqui  onde voc armazena suas imagens do Docker para que possam ser executadas pelo ECS ou Fargate.

    

Serverless

O que  serverless?

     Serverless  um novo paradigma no qual os desenvolvedores no precisam mais gerenciar servidores...
     Eles apenas implantam cdigo
     Eles apenas implantam... funes!
     Inicialmente... Serverless == FaaS (Funo como um Servio)
     Serverless foi pioneirizado pelo AWS Lambda, mas agora tambm inclui qualquer coisa que seja gerenciada: "banco de dados, mensagens, armazenamento, etc."
     Serverless no significa que no existam servidores...
    Significa apenas que voc no os gerencia / provisiona / v.

Lambda

Por que usar AWS Lambda?



Benefcios do AWS Lambda

     Preos Simples:
     Pague por solicitao e tempo de computao
     Camada gratuita de 1.000.000 de solicitaes do AWS Lambda e 400.000 GBs de tempo de computao
     Integrado com todo o conjunto de servios da AWS
     Orientado a Eventos: funes so invocadas pela AWS quando necessrio
     Integrado com muitas linguagens de programao
     Monitoramento fcil atravs do AWS CloudWatch
     Fcil para obter mais recursos por funo (at 10GB de RAM!)
     Aumentar a RAM tambm melhorar a CPU e a rede!

Suporte de linguagens para AWS Lambda

     Node.js (JavaScript)
     Python
     Java (compatvel com Java 8)
     C# (.NET Core)
     Golang
     C# / Powershell
     Ruby
     API de Tempo de Execuo Personalizada (suportada pela comunidade, exemplo: Rust)
     Imagem de Continer Lambda
     A imagem do continer deve implementar a API de Tempo de Execuo do Lambda
     ECS / Fargate so preferidos para executar imagens Docker arbitrrias

Preos do AWS Lambda: exemplo

     Voc pode encontrar informaes gerais de preos aqui: https://aws.amazon.com/lambda/pricing/
     Pagamento por chamadas:
     Os primeiros 1.000.000 de solicitaes so gratuitos
     $0,20 por 1 milho de solicitaes aps isso ($0,0000002 por solicitao)
     Pagamento por durao: (em incrementos de 1 ms)
     400.000 GB-segundos de tempo de computao por ms gratuitos
     == 400.000 segundos se a funo tiver 1GB de RAM
     == 3.200.000 segundos se a funo tiver 128 MB de RAM
     Depois disso, $1,00 por 600.000 GB-segundos
     Geralmente  muito barato rodar o AWS Lambda, por isso  muito popular.

AWS Batch    

     Processamento em lote totalmente gerenciado em qualquer escala
     Execute eficientemente 100.000s de trabalhos de computao em lote na AWS
     Um trabalho em "lote"  um trabalho com um incio e um fim (oposto a contnuo)
     O Batch lanar dinamicamente instncias EC2 ou Instncias Spot
     O AWS Batch provisiona a quantidade certa de computao / memria
     Voc submete ou agenda trabalhos em lote e o AWS Batch faz o resto!
     Trabalhos em lote so definidos como imagens Docker e so executados no ECS
     til para otimizaes de custos e foco menor na infraestrutura

    

Batch vs Lambda

 Lambda:

     Limite de tempo
     Runtimes limitados
     Espao de disco temporrio limitado
     Sem servidor

 Batch:

     Sem limite de tempo
     Qualquer runtime, desde que seja empacotado como uma imagem Docker
     Dependente do EBS / armazenamento da instncia para espao em disco
     Dependente do EC2 (pode ser gerenciado pela AWS)


Amazon Lightsail    

     Servidores virtuais, armazenamento, bancos de dados e rede
     Preos baixos e previsveis
     Alternativa mais simples ao uso de EC2, RDS, ELB, EBS, Route 53...
     timo para pessoas com pouca experincia na nuvem!
     Pode configurar notificaes e monitoramento dos seus recursos do Lightsail
     Casos de uso:
     Aplicaes web simples (possui modelos para LAMP, Nginx, MEAN, Node.js...)
     Websites (modelos para WordPress, Magento, Plesk, Joomla)
     Ambiente de desenvolvimento / teste
     Tem alta disponibilidade, mas sem autoescalonamento, integraes limitadas com a AWS


Outros Servios de Computao - Resumo

     Docker: tecnologia de continer para executar aplicativos
     ECS: executa contineres Docker em instncias EC2
     Fargate:
     Executa contineres Docker sem provisionar a infraestrutura
     Oferta sem servidor (sem instncias EC2)
     ECR: Repositrio de Imagens Docker Privado
     Batch: executa trabalhos em lote na AWS em instncias EC2 gerenciadas
     Lightsail: preos previsveis e baixos para pilhas de aplicativos e bancos de dados simples.

Resumo do Lambda

     Lambda  sem servidor, Function as a Service, escala contnua, reativo
     Faturamento do Lambda:
     Pelo tempo de execuo multiplicado pela RAM provisionada
     Pelo nmero de invocaes
     Suporte de Linguagem: muitas linguagens de programao, exceto (arbitrariamente) Docker
     Tempo de Invocao: at 15 minutos
     Casos de Uso:
     Criar miniaturas para imagens enviadas para o S3
     Executar uma tarefa cron sem servidor
     API Gateway: expor funes Lambda como API HTTP
