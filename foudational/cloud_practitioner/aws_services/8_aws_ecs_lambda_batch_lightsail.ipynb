Docker

O que é o Docker?

    • Docker é uma plataforma de desenvolvimento de software para implantar aplicativos
    • Aplicativos são empacotados em contêineres que podem ser executados em qualquer sistema       operacional
    • Aplicativos funcionam da mesma forma, independentemente de onde são executados
    • Qualquer máquina
    • Sem problemas de compatibilidade
    • Comportamento previsível
    • Menos trabalho
    • Mais fácil de manter e implantar
    • Funciona com qualquer linguagem, qualquer sistema operacional, qualquer tecnologia
    • Escala contêineres para cima e para baixo muito rapidamente (segundos)

    

    
Onde as imagens do Docker são armazenadas?

    • As imagens do Docker são armazenadas em Repositórios do Docker
    • Público: Docker Hub https://hub.docker.com/
    • Encontre imagens base para muitas tecnologias ou sistemas operacionais:
    • Ubuntu
    • MySQL
    • NodeJS, Java...
    • Privado: Amazon ECR (Elastic Container Registry)

Docker versus Máquinas Virtuais

    • Docker é "quase" uma tecnologia de virtualização, mas não exatamente
    • Recursos são compartilhados com o host => muitos contêineres em um servidor



ECS

    ECS = Elastic Container Service
    • Lança contêineres Docker na AWS
    • Você precisa provisionar e manter a infraestrutura (as instâncias EC2)
    • A AWS cuida de iniciar / parar os contêineres
    • Possui integrações com o Application Load Balancer



Fargate

    Launch Docker containers on AWS
    • Você não provisiona a infraestrutura (sem instâncias EC2 para gerenciar) - mais simples!
    • Oferta serverless
    • A AWS apenas executa os contêineres para você com base na CPU / RAM que você precisa



ECR

    Elastic Container Registry
    • Registro privado do Docker na AWS
    • Aqui é onde você armazena suas imagens do Docker para que possam ser executadas pelo ECS ou Fargate.

    

Serverless

O que é serverless?

    • Serverless é um novo paradigma no qual os desenvolvedores não precisam mais gerenciar servidores...
    • Eles apenas implantam código
    • Eles apenas implantam... funções!
    • Inicialmente... Serverless == FaaS (Função como um Serviço)
    • Serverless foi pioneirizado pelo AWS Lambda, mas agora também inclui qualquer coisa que seja gerenciada: "banco de dados, mensagens, armazenamento, etc."
    • Serverless não significa que não existam servidores...
    Significa apenas que você não os gerencia / provisiona / vê.

Lambda

Por que usar AWS Lambda?



Benefícios do AWS Lambda

    • Preços Simples:
    • Pague por solicitação e tempo de computação
    • Camada gratuita de 1.000.000 de solicitações do AWS Lambda e 400.000 GBs de tempo de computação
    • Integrado com todo o conjunto de serviços da AWS
    • Orientado a Eventos: funções são invocadas pela AWS quando necessário
    • Integrado com muitas linguagens de programação
    • Monitoramento fácil através do AWS CloudWatch
    • Fácil para obter mais recursos por função (até 10GB de RAM!)
    • Aumentar a RAM também melhorará a CPU e a rede!

Suporte de linguagens para AWS Lambda

    • Node.js (JavaScript)
    • Python
    • Java (compatível com Java 8)
    • C# (.NET Core)
    • Golang
    • C# / Powershell
    • Ruby
    • API de Tempo de Execução Personalizada (suportada pela comunidade, exemplo: Rust)
    • Imagem de Contêiner Lambda
    • A imagem do contêiner deve implementar a API de Tempo de Execução do Lambda
    • ECS / Fargate são preferidos para executar imagens Docker arbitrárias

Preços do AWS Lambda: exemplo

    • Você pode encontrar informações gerais de preços aqui: https://aws.amazon.com/lambda/pricing/
    • Pagamento por chamadas:
    • Os primeiros 1.000.000 de solicitações são gratuitos
    • $0,20 por 1 milhão de solicitações após isso ($0,0000002 por solicitação)
    • Pagamento por duração: (em incrementos de 1 ms)
    • 400.000 GB-segundos de tempo de computação por mês gratuitos
    • == 400.000 segundos se a função tiver 1GB de RAM
    • == 3.200.000 segundos se a função tiver 128 MB de RAM
    • Depois disso, $1,00 por 600.000 GB-segundos
    • Geralmente é muito barato rodar o AWS Lambda, por isso é muito popular.

    • Sem servidor e escalável
    • Suporta APIs RESTful e WebSocket

AWS Batch    

    • Processamento em lote totalmente gerenciado em qualquer escala
    • Execute eficientemente 100.000s de trabalhos de computação em lote na AWS
    • Um trabalho em "lote" é um trabalho com um início e um fim (oposto a contínuo)
    • O Batch lançará dinamicamente instâncias EC2 ou Instâncias Spot
    • O AWS Batch provisiona a quantidade certa de computação / memória
    • Você submete ou agenda trabalhos em lote e o AWS Batch faz o resto!
    • Trabalhos em lote são definidos como imagens Docker e são executados no ECS
    • Útil para otimizações de custos e foco menor na infraestrutura

    

Batch vs Lambda

• Lambda:

    • Limite de tempo
    • Runtimes limitados
    • Espaço de disco temporário limitado
    • Sem servidor

• Batch:

    • Sem limite de tempo
    • Qualquer runtime, desde que seja empacotado como uma imagem Docker
    • Dependente do EBS / armazenamento da instância para espaço em disco
    • Dependente do EC2 (pode ser gerenciado pela AWS)


Amazon Lightsail    

    • Servidores virtuais, armazenamento, bancos de dados e rede
    • Preços baixos e previsíveis
    • Alternativa mais simples ao uso de EC2, RDS, ELB, EBS, Route 53...
    • Ótimo para pessoas com pouca experiência na nuvem!
    • Pode configurar notificações e monitoramento dos seus recursos do Lightsail
    • Casos de uso:
    • Aplicações web simples (possui modelos para LAMP, Nginx, MEAN, Node.js...)
    • Websites (modelos para WordPress, Magento, Plesk, Joomla)
    • Ambiente de desenvolvimento / teste
    • Tem alta disponibilidade, mas sem autoescalonamento, integrações limitadas com a AWS


Outros Serviços de Computação - Resumo

    • Docker: tecnologia de contêiner para executar aplicativos
    • ECS: executa contêineres Docker em instâncias EC2
    • Fargate:
    • Executa contêineres Docker sem provisionar a infraestrutura
    • Oferta sem servidor (sem instâncias EC2)
    • ECR: Repositório de Imagens Docker Privado
    • Batch: executa trabalhos em lote na AWS em instâncias EC2 gerenciadas
    • Lightsail: preços previsíveis e baixos para pilhas de aplicativos e bancos de dados simples.

Resumo do Lambda

    • Lambda é sem servidor, Function as a Service, escala contínua, reativo
    • Faturamento do Lambda:
    • Pelo tempo de execução multiplicado pela RAM provisionada
    • Pelo número de invocações
    • Suporte de Linguagem: muitas linguagens de programação, exceto (arbitrariamente) Docker
    • Tempo de Invocação: até 15 minutos
    • Casos de Uso:
    • Criar miniaturas para imagens enviadas para o S3
    • Executar uma tarefa cron sem servidor
    • API Gateway: expor funções Lambda como API HTTP
