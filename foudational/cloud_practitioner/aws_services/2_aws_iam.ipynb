IAM User & Groups

    Não utilize a conta root para operar a AWS, crie um usuário apra usar a AWS

    O ideal é sempre deixar os usuários com o menor número de permissoes possível

    Usuários possuem acesso global

    Se aparecer Account ID abaixo do nome, significa que você está na conta root.

    Criando um usuário:

    Primeiro você cadastra o nome dele
    Depois você escolhe como será gerada a sennha de acesso dele

    Agora você seleciona as permissões desse usuário:

    Significa que você terá escolher um grupo, que contem as permissões que você deseja que o usuário X tenha,se ainda não exista nenhum grupo as informações especificas,você pega e cria um novo grupo e seleciona as permissões que você deseja.

    Após fazer os passos acima, você verifica se todas as informações que você optou estão corretas.

    Nessa estapa você poderá criar TAGS, elas servem, para ajudar nos seus LOGS, pq você pode escrever assim:

    Key: Departamento
    Value: Engenharia de Dados

    lembrando que acima é um exemplo, mas é algo opcional e que é legal de ter, pois trás mais clareza aos processos.

    Depois de ter criado o usuário, faça o login na AWS através do IAM User. 

    OBS: se for sua primeira vez na AWS, crie seu ALIAS, isso mascara o número na sua URL do root, aumentando o seu nível de proteção.

IAM Policies inheritance

    Um usuário pode participar de vários grupos
    Cada grupo pode pussuir várias permiçõees

    EX: o usuários cristian, pertence ao grupo ACESSO_GERAL_AWS, esse grupo chamado ACESSO_GERAL_AWS, possivelmente possui a permissão de fazer tudo dentro da AWS, permitindo usar todos os 403(dia 18/1/23) serviços da AWS.
    Isso pode mudar de permissão para permissão, tem cenários que o usuário só poderra ler arquivos. Isso pode variar muito, pois, como dito, devemos limitar ao máximo as permissões dos usuários, deixando acesso ao necessários para que o usuário faça seu trabalho. Dessa forma, protegemos ao máximo nosso sistema de cloud.

    Você poderia criar permissões com as necessidades que você julga necessário. Você pode fazer isso de uma interface ou criando um json no padrão das permissões da AWS.
    Lembrando que * significa tudo, ou seja, quando você coloca * em alguma coisa, sigifica que você está dizendo tudo.

IAM Password Policy

    Basicamente temos que ter o MFA, autenticação de 2 fatores, isso ajude e muita a aumentar a proteção da nossa conta. Existem dispositivos online e aparelhos físicos... Vai da sua necessidade/desejo de optar pela sua verificação de 2 fatores...

    Se você for na aba lateral e clicar em Account Settings, verá que você poderá optar por mudar o padrão das suas senhas, por padrão usamos o padrão IAM User, mas se você clicar em custom, verá que pode definar varias regras para gerar e manter uma senha de usuário da na AWS.

    Ativando seu MFA:

    Por padrão você não terá, entao você precisa clicar no seu nome, canto superior direito e selecioanr a opção SECURITY CREDENTIALS, de cara você verá uma mensagem dizendo que você não possui verificação de 2 fatores.

    Para criar o MFA device

    Clique em Assing MFA

    Depois escreva o nome do seu MFA e selecione o modelo de MFA que você deseja.

    Quanto você fizer isso, você entrará em uma tela que mostra que você precisa cadastrar 2 chaves. 
    Para faze isso, você precisa escanear o QR code com seu APP ou mecanismo de MFA account.

    No caso de APP:

    Quando você escanear, o seu aplicativo irá lhe fornecerá tokens, você precisa cadastrar esses números/tokens na parte de baixo, a AWS exigirá 2 tokens. Após fazer isso, você avança e pronto.
    Voce criou seu MFA e sua conta está mais segura.

IAM Rules

    São regras de segurança para serviços criados na AWS
    EX: Ec2 Lambda Functions

    Como criar uma role:
    Vá no IAM - menu
    E selecione Roles - Create Role (em larnaja)

    Selecione Trusted entity type
    AWS Service deve ser o mais comum

    depois selecione o caso de usso em Use Case
    ex: EC2
    Clique em Next

    Selecione a permissão que você deseja

    Após, coloque o nome da Regra e coloque a descrição da role, pra que ela serve o que deve fazer, etc..... Não é obrigatório, mas é bom...
