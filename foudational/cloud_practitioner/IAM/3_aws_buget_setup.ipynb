AWS Budget Setup

Depois de configurar as senhas e permissões do da AWS, você precisa configurar os alarmes de despesar, para evitar possíveis surpresas....

Por padrão só a ROOT ACCOUNT poderá visualizar a área de Billing and Cost Management
Para permitir que outras contas acessem a área de Billing and Cost Management, você precisa ir na usa conta Root, clicar no canto superior em Account, e descer até a parte do IAM user and role access to Billing information. Essa parte estará desmarcada. Voce precisa marcar ela, fazendo isso, você permitirá que outras contas tenham acessam ao Billing and Cost Management.

Após fazer isso, você pode ir em Billing and Payments -> Bills e visualizar o custo da sua infraestrutura.

Caso você seja Free Tier, você pode ir em Cost Analysis - Free Tier -> e visulizar todos os serviços Free e quanto você usou até o momento.

PARA EVITAR SURPRESAS

Crie alarmes de gastos.

Siga esse caminho: Billing and Cost Management > Budgets > Overview

Clique em Creat a Budget e crie seu aviso de custos, dessa forma você não terá surpresas no final do mês!!!
