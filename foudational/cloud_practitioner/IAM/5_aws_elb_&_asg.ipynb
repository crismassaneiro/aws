Scalability & High Availability

Scalabilidade significa que sua aplicação/sistema, pode lidar com cargas maiores se adaptando

Existe 2 tipos de escalabilidade:
    Vertical
    Horizontal
    
Escalablidade Vertical

Pensa assim, você tem 1 EC2 podre e então para aumentar seu poder computacional você faz um upgrade para um EC2 superior

Mas querendo ou não, você chegará em um limite de instância, porque, por mais que sua máquina seja a mais forte da AWS, em algum momento ela chegará no limite, dependendo do seu cenário. Fora questões de custo. 
Muito comum em data bases, isso ocorre pq o tempo de processamento aumenta conforme a base cresce, então você acaba precisando de mais poder computacional para rodar o código

Escalabilidade Horizontal


Você possui um EC2 e precisa aumentar o poder computacional, então você cria outra intância EC2, identica, se precisar, criar outra e outra, até suprir sua necessidade.
