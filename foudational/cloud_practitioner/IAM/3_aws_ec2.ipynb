AWS EC2 RESUME

EC2 é um dos serviços mais populares da AWS

EC2 = Elastic Compute Cloud = infrastructure as a service

Em resumo, permite:
    Criar maquinas virtuais (EC2)
    Guardar dados em drivers vitural (EBS)
    Distribuir carregamentos entre maquinas (ELB)
    Dimencionar serviços usando o auto scaling group (ASG)    

Conhecer sobre o EC2 é fundamental para entender como a Cloud Funciona

EC2 User data
São comandos que você pode criar através de um scrip, para que, quando sua máquina for startar pela primeira vez ela siga.
    Tipos de comandos:
    Instalar atualizações
    Instalar softwares
    Fazer o download de arquivos da internet
    Tudo o que você quiser....

Existem várias configurações máquinas, então você sempre terá alguma máquina que conseguirá rodar seus códigos.

Como lançar uma instância   EC2

Name and tags

    Selecione sua instancia

Security Grups 

São a base da segurança da AWS
São responsáveis por controlar o que pode sair e o que pode entrar nas instancias Ec2
Só possuem regras de permissão
Eles são a proteção dos nossos Ec2
Eles regulam:

    Acesso a portas

    IPs ranges IPv4 E IPv6

    Controlam a entrada

    Controlam a saída

Importante saber:

Você pode usar um Security Group para várias instâncias
Uma instancia pode ter vários Secutiry Groups
Security Grups funcionam por região/VPC, se você mudar, precisa criar outro Security Group para região
Security Groups não estão dentro do EC2, eles funcionam a parte 
É muito recomendável manter um Security Group separado para SSH access
Se a sua aplicação não ficar acessível (time out), por causa de um erro no seu Security Group
Se aparecer o erro CONEXÃO RECUSA, significa que é um arro na aplicaçaõ ou ela não foi lançada.
Todo trafico inboud é bloqueado por padrão
Todo trafico outbound é autorizado por padrão

Portas que devemos saber para a prova:

22 = SSH(Secure Shell) - log into a Linux instance
21 = FTP(File Transfer Protocol - upload files into a file share)
22 = SFTP(Secure File Transfer Protocol) - upload files using SSH
80 = HTTPS - access ansecured websites
443 = HTTPS - access secured websites
3389 = RDP(Remote Desktop Protocol) - log into a Windows instance

Sobre o SSH:



Existe duas formas de se conectar com sua instância SSH

Uma delas é através do seu Desktop, você precisará da sua chave.pem que você faz o download quando criar uma instancia Ec2.

Outras forma é você se conectar diratamente pelo console, indo em Intances - Ec2: selecione a opção conect

Lembrse que você terá de ter configura o Security Group para conseguir se conectar.

Nunca salve suas chaves access key e security key dentro da sua instancia ec2
IAM

Para que suas intancias Ec2 tenhama cesso a sua chaves, você precisa adicionar IAM Role em sua instancia EC2. 
Para fazer isso vá até sua intancia, selecione ele, clique em actions, security e Modify IAM Role fazendo isso, você poderá escolher uma iam Role ou criar uma.
