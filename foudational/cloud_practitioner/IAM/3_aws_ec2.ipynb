S possuem regras de permisso
Eles so a proteo dos nossos Ec2
Eles regulam:

    Acesso a portas

Importante saber:

Voc pode usar um Security Group para vrias instncias
Uma instancia pode ter vrios Secutiry Groups
Security Grups funcionam por regio/VPC, se voc mudar, precisa criar outro Security Group para regio
Security Groups no esto dentro do EC2, eles funcionam a parte 
 muito recomendvel manter um Security Group separado para SSH access
Se a sua aplicao no ficar acessvel (time out), por causa de um erro no seu Security Group
Se aparecer o erro CONEXO RECUSA, significa que  um arro na aplicaa ou ela no foi lanada.
Todo trafico inboud  bloqueado por padro
Todo trafico outbound  autorizado por padro

Portas que devemos saber para a prova:

22 = SSH(Secure Shell) - log into a Linux instance
21 = FTP(File Transfer Protocol - upload files into a file share)
22 = SFTP(Secure File Transfer Protocol) - upload files using SSH
80 = HTTPS - access ansecured websites
443 = HTTPS - access secured websites
3389 = RDP(Remote Desktop Protocol) - log into a Windows instance
