### Passo 3 - Configure seu perfil CLI

#### Digite no cmd que você abriu:
- digite:  aws configure --profile SEUS_NOME_AWS (escreva SEU_NOME e coloque _AWS, para você lembrar do que se trata esse nome, mas se quiser escrever qualquer nome, OK : ), é o seu pefil, não é mesmo?)

#### Isso iniciará um assistente interativo que solicitará as seguintes informações:

- AWS Access Key ID: Insira a sua Access Key da AWS. (essa informação é gerada no IAM da sua conta AWS, isso gera um excel com AcCess Key e Secrect Key) OBS: se você perdeu esse excel, precisirá exlcuir as chaves existentes e criar novas, para poder gerar um excel com informações novas. !!Não perca o EXECEL, isso pode dar muita merda se cair em mãos erradas.

     -> depois de digitar, clique ENTER
- AWS Secret Access Key: Insira a sua Secret Key da AWS. (essa informação é gerada no IAM da sua conta AWS, isso gera um excel com ACcess Key e Secrect Key) OBS: se você perdeu esse excel, precisirá exlcuir as chaves existentes e criar novas, para poder gerar um excel com informações novas. !!Não perca o EXECEL, isso pode dar muita merda se cair em mãos erradas.

     -> depois de digitar, clique ENTER

- Default region name: Insira a região AWS padrão que você deseja usar. Por exemplo, "us-east-1" para a região leste dos EUA.

     -> depois de digitar, clique ENTER
- Default output format: O formato de saída padrão, que normalmente é "json". Você pode aceitar o padrão ou escolher outro formato, se desejar.

     apenas clique ENTER (configurar isso não é necessário, o default será JSON, por isso é só clicar ENTER)

    #### Certifique-se de que as credenciais inseridas estejam corretas e tenham as permissões adequadas para as operações que você planeja realizar com o AWS CLI.

    #### FEITO!!! Você configurou seu PERFIL : )
